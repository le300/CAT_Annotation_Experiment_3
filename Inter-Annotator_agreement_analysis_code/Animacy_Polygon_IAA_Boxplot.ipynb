```python
# BOXPLOTs - Function for Boxplot Visualization of pairs of boxplots 

# code adapted from https://stackoverflow.com/questions/16592222/matplotlib-group-boxplots

def create_boxplots(pairwise_data1, pairwise_data2, ticks, label1, label2, title, save_fig):
    """
    Takes: pairwise_data1 - a list of lists, organized in lists per category.
        pairwise_data2 - a list of lists, organized in lists per category.
        ticks - list of strings which serve as category labels on the x-axis.
        label1 - string label for red boxes (pairwise_detail1 values).
        label2 - string label for blue boxes (pairwise_detail2 values). 
        title - string title for whole boxplot.
        save_fig - a boolean, if true then the figure is saved as a jpeg, and the figure is not saved if false.
    Returns: no return, but a boxplot is created. 
    """
    
    def set_box_color(bp, color):
        plt.setp(bp['boxes'], color=color)
        plt.setp(bp['whiskers'], color=color)
        plt.setp(bp['caps'], color=color)
        plt.setp(bp['medians'], color=color)

    fig = plt.figure(1)
    ax = fig.add_subplot(111,aspect='auto')
    
    meanpointprops_blue = dict(marker='o', markeredgecolor='#2C7BB6',
                      markerfacecolor='#2C7BB6')

    meanpointprops_red = dict(marker='o', markeredgecolor='#D7191C',
                      markerfacecolor='#D7191C')

    bpl = plt.boxplot(pairwise_data1, positions=np.array(range(len(pairwise_data1)))*2.0-0.4, sym='', widths=0.6, showmeans=True, meanprops=meanpointprops_red)
    bpr = plt.boxplot(pairwise_data2, positions=np.array(range(len(pairwise_data2)))*2.0+0.4, sym='', widths=0.6, showmeans=True, meanprops=meanpointprops_blue)
    set_box_color(bpl, '#D7191C') # colors are from http://colorbrewer2.org/
    set_box_color(bpr, '#2C7BB6')

    # draw temporary red and blue lines and use them to create a legend
    plt.plot([], c='#D7191C', label=label1)
    plt.plot([], c='#2C7BB6', label=label2)
    plt.plot([], c='white', marker='o', markerfacecolor="black", label="Mean")
    plt.plot([], c='white', label="n=10 for each story")
    plt.legend()
    tick_positions = [x for x in list(range(0, len(ticks) * 2, 2))]
    print(tick_positions)
    plt.xticks(tick_positions, ticks)
    ax.set_xticklabels(ticks, ha="center", rotation_mode='anchor')
    #ax.set_xlabel(" n=10                           n=10                                              n=10                         n=10 ", fontsize=8)
    ax.set_title(title, fontsize=10)
    plt.xlim(-2, len(ticks)*2)
    plt.ylim(-0, 1)
    plt.grid()
    plt.tight_layout()
    
    plt.rcParams["figure.figsize"] = (4,4)
    
    if save_fig == True:
        plt.savefig('saved_fig.jpeg', dpi=300, bbox_inches = "tight")
    
```
